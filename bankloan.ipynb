```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_excel('/content/sample_data/samplebankloandata.xlsx')
print(df.head())
print(df.tail())
print(df.isnull().sum())
print(df.describe())
```

```python
print(df.columns)
```

```python
sns.histplot(data=df,x='annual_inc',kde=True)
plt.title("Annual income")
plt.show()
```

```python
sns.histplot(data=df,x='loan_amount',kde=True)
plt.title("loan amount")
plt.show()
```

```python
plt.pie(df['home_ownership'].value_counts(),labels=df['home_ownership'].value_counts().index,autopct='%1.1f%%')
plt.title('HOME_OWNERSHIP')
plt.show()
```

```python
sns.scatterplot(data=df,x='annual_inc',y='loan_amount')
plt.title('Annual Income vs Loan Amount')
plt.show()
```

```python
# Calculate mean, median, and mode of annual_inc
mean_annual_inc = df['annual_inc'].mean()
median_annual_inc = df['annual_inc'].median()
mode_annual_inc = df['annual_inc'].mode()

print("Mean annual income:", mean_annual_inc)
print("Median annual income:", median_annual_inc)
print("Mode annual income:", mode_annual_inc)
```

```python
# Calculate mean, median, and mode of interest_rate
mean_interest_rate = df['interest_rate'].mean()
median_interest_rate = df['interest_rate'].median()
mode_interest_rate = df['interest_rate'].mode()

print("Mean interest rate:", mean_interest_rate)
print("Median interest rate:", median_interest_rate)
print("Mode interest rate:", mode_interest_rate)
```

```python
# Calculate the total sum of total_pymnt
total_payment_received = df['total_pymnt'].sum()
print("Total payment received:", total_payment_received)
```

```python
# Calculate the percentage of loans that are fully paid off
fully_paid_percentage = (df['loan_condition'] == 'Fully Paid').mean() * 100
print("Percentage of loans fully paid off:", fully_paid_percentage)
```

```python
# Calculate the average interest payment per loan
average_interest_payment = df['total_rec_prncp'].mean()

print("Average interest payment per loan:", average_interest_payment)
```

```python
# Count the frequency of each loan purpose
loan_purpose_counts = df['purpose'].value_counts()
plt.figure(figsize=(10, 6))
loan_purpose_counts.plot(kind='bar', color='skyblue')
plt.title('Distribution of Loan Purposes')
plt.xlabel('Loan Purpose')
plt.ylabel('Frequency')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
